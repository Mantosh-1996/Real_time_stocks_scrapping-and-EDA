```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Chrome(r"C:\Users\manto\.wdm\drivers\chromedriver\win32\94.0.4606.61\chromedriver.exe")
```

```python
driver.get("https://www.moneycontrol.com/markets/indian-indices/")
driver.find_element(By.XPATH, "//*[@id='indices-2']/a").click()
```

```python
driver.find_element(By.XPATH, "//*[@id='con_comn']/li[3]/h2/a").click()
```

**Extracting Stocks of Automobile sectors**

```python

x= driver.find_element(By.XPATH, "//*[@id='indicesTable']")
tx= x.text
l=tx.split('\n')
l[0] = ' '.join(l[:10])
del l[1:10]
l
d=[]
for item in l:
    item = item.split(' ')
    d.append(item)


for item in d:
    
    if item[1].isalpha() and item[2].isalpha() and item[3].isalpha():
        item [0] = item[0]+ ' ' + item[1] + ' '+ item[2]+ ' '+item[3]
        del item[1:4]
    elif item[1].isalpha() and item[2].isalpha():
        item [0] = item[0]+ ' ' + item[1] + ' '+ item[2]
        del item[1:3]
    elif item[1].isalpha():
        item [0] = item[0]+ ' ' + item[1]
        del item[1]
d[0][0]= 'Name'
d[0][1] = "LTP"
d[0][2] = 'P/E'
d[0][3] = "Debt_to_equity"
d[0][4] = "ESP"
d[0][5] = "BVPS"
d[0][6] = "Net_Profit(RS.cr)"
d[0][7] = "DSP(RS.)"
d[0][8] = "NPM%"
d[0][9] = "ROE%"
del d[0][10:]
Automobile= pd.DataFrame(d)

Automobile.to_csv("Automobile.CSV")
Automobile.head()
```
